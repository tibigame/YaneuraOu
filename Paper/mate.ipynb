```python
import numpy as np

report_nodes = 1000000

class MateSfen:
    def __init__(self, split_line):
        if len(split_line) <= 2:
            self.sfen = split_line[0]
            self.mate = split_line[1]
            self.nodes = 0
            return
        self.sfen = split_line[0]
        self.mate = split_line[1]
        self.nodes = int(split_line[2])

def get_nodes_list(class_list):
    return np.sort(np.array([int(result.nodes) for result in class_list]))[::-1]

result_list = []       

with open("result.txt") as f:
    for line in f:
        if len(line) > 10:
            result_list.append(MateSfen(line.split(",")))

result_list = sorted(result_list, key=lambda t: t.nodes)

e_list = [result for result in result_list if result.mate == "e"]
mate_list = [result for result in result_list if result.mate == "m"]
nmate_list = [result for result in result_list if result.mate == "n"]
unknown_list = [result for result in result_list if result.mate == "k"]
None
```

```python
print(f"結果の表示")
print(len(result_list))
print(f"詰み: {len(mate_list)}")
print(f"不明: {len(unknown_list)}")
print(f"不詰み: {len(nmate_list)}")
print(f"エラー: {len(e_list)}")
print(f"エラーの局面：")
[print(f"sfen {result.sfen}") for result in e_list]
print(f"詰みの局面： nodes={report_nodes}以上")
[print(f"sfen {result.sfen}, nodes={result.nodes}") for result in mate_list if result.nodes >= report_nodes]
print(f"不明の局面： nodes={report_nodes}以上")
[print(f"sfen {result.sfen}, nodes={result.nodes}") for result in unknown_list if result.nodes >= report_nodes]
print(f"不詰みの局面： nodes={report_nodes}以上")
[print(f"sfen {result.sfen}, nodes={result.nodes}") for result in nmate_list if result.nodes >= report_nodes]
None
```

```python
def describe_list(list):
    def print_ex(str, num):
        print(f"{str}は {num} です。")
    
    print_ex("　最小値", np.min(list))
    print_ex("　　平均", np.mean(list))
    print_ex("　中央値", np.median(list))
    print_ex("　最大値", np.max(list))
    print_ex("標準偏差", np.std(list))

print(f"統計情報")
print(f"---------- 全体 ----------")
describe_list(get_nodes_list(result_list))
print(f"---------- 詰み ----------")
describe_list(get_nodes_list(mate_list))
print(f"---------- 不明 ----------")
describe_list(get_nodes_list(unknown_list))
print(f"---------- 不詰み --------")
describe_list(get_nodes_list(nmate_list))
```
