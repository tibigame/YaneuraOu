```python
import numpy as np
import scipy

class MateSfen:
    def __init__(self, split_line):
        if len(split_line) <= 2:
            self.sfen = split_line[0]
            self.mate = split_line[1]
            self.nodes = 0
            return
        self.sfen = split_line[0]
        self.mate = split_line[1]
        self.nodes = split_line[2]

result_list = []       

with open("result.txt") as f:
    for line in f:
        if len(line) > 10:
            result_list.append(MateSfen(line.split(",")))

nodes = np.sort(np.array([int(result.nodes) for result in result_list]))[::-1]

def print_ex(str, num):
    print(f"{str}は {num} です。")

def describe_list(list):
    print_ex("　最小値", np.min(list))
    print_ex("　　平均", np.mean(list))
    print_ex("　中央値", np.median(list))
    print_ex("　最大値", np.max(list))
    print_ex("標準偏差", np.std(list))
            
print(len(result_list))
print(f"詰み: {[result.mate for result in result_list].count('m')}")
print(f"不明: {[result.mate for result in result_list].count('k')}")
print(f"不詰み: {[result.mate for result in result_list].count('n')}")
describe_list(nodes)
print(f"nodes: {nodes})")
```
